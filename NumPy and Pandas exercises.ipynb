##### 10.Append values to the end of an array using np.append().
- Original array: [10, 20, 30]
- Expected Output: [10 20 30 40 50 60 70 80 90]

```python
array = [10,20,30]
```

```python
values = [40,50,60,70,80,90]
```

```python
array
```

```python
for x in values:
    array = np.append(array,x)
```

```python
array
```
